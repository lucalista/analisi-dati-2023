# Generatore picchi

Questo script genera campioni di segnale più fondo simulati.

**Non è stato fornito agli studenti prima della lezione. Sono stati forniti solo i dataset chiedendo di fare dei fit con diversi modelli di segnale e fondo**

```python
from matplotlib import pyplot as plt
import numpy as np
import math

np.random.seed(12345)
```

```python
def plot_data(n_bins=50, x_max=50, filename="data/fit-data.txt"):
    fig, ax = plt.subplots(figsize=(9,6),dpi=100)
    x_range = (0, x_max)
    data_y, bins, patches = ax.hist(evt_x, n_bins, range=x_range, density=False, 
        histtype=u'step', edgecolor='b', linewidth=2, label='histogram')
    ax.hist(evt_x, n_bins, range=x_range, density=False, color='b', alpha=0.2)
    ax.scatter(evt_x, np.full_like(evt_x, 0.5), marker='o', color='r', s=10, alpha=0.2, linewidth=0, label='entries')
    plt.xlabel('x')
    plt.ylabel('$\Delta N/\Delta x$')
    plt.xlim(0, x_max)
    plt.legend()
    plt.show()
    np.savetxt(filename, evt_x)
```

Singolo picco su fondo esponenziale

```python
nexp_sig, nexp_bkg1, nexp_bkg2 = 90, 300, 50
l = 12
mu, sigma = 10, 1.5
n_sig = np.random.poisson(nexp_sig)
n_bkg1 = np.random.poisson(nexp_bkg1)
n_bkg2 = np.random.poisson(nexp_bkg2)
sig_x = np.array([np.random.normal(loc=mu, scale=sigma) for i in range(n_sig)])
bkg1_x = np.array([np.random.exponential(scale=l) for i in range(n_bkg1)])
bkg2_x = np.array([np.random.uniform(0, 100) for i in range(n_bkg2)])
evt_x = np.concatenate([sig_x, bkg1_x, bkg2_x])
np.random.shuffle(evt_x)
plot_data(100, 100, filename="data/fit-data-1.txt")
```

Due picchi, ben separati.

```python
nexp_sig1, nexp_sig2, nexp_bkg1, nexp_bkg2 = 90, 60, 300, 280
l = 25
mu1, sigma1, mu2, sigma2 = 16, 1.5, 34, 2.3
n_sig1, n_sig2 = np.random.poisson(nexp_sig1), np.random.poisson(nexp_sig2)
n_bkg1 = np.random.poisson(nexp_bkg1)
n_bkg2 = np.random.poisson(nexp_bkg2)
sig1_x = np.array([np.random.normal(loc=mu1, scale=sigma1) for i in range(n_sig1)])
sig2_x = np.array([np.random.normal(loc=mu2, scale=sigma2) for i in range(n_sig2)])
bkg1_x = np.array([np.random.exponential(scale=l) for i in range(n_bkg1)])
bkg2_x = np.array([np.random.uniform(0, 150) for i in range(n_bkg2)])
evt_x = np.concatenate([sig1_x, sig2_x, bkg1_x, bkg2_x])
evt_x = evt_x[evt_x<150]
np.random.shuffle(evt_x)
plot_data(150, 150, filename="data/fit-data-2.txt")
```
